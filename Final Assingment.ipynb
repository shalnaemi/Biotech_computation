```python
# This code was generated using AI. Prompt used: "Write a Python script that reads all .txt files from a directory, extracts tab-separated column headers from the first file, skips empty lines, and stores the remaining data lines in a list"
import os

col_names = [] # stores column names from the first file
data_lines = [] # stores all data

directory = "/Users/shaikhaalnaemi/Library/CloudStorage/OneDrive-UniversityofAberdeen/Biotechnology/Python" # path to make a directory

# To loop through all txt files in the directory
for filename in os.listdir(directory):
    file_path = os.path. join(directory, filename)

     # To check if it is a file and ends with .txt
    if os.path.isfile(file_path) and filename.endswith(".txt"):
        with open (file_path, "r") as myfile:
            for idx, line in enumerate(myfile):
                line = line.strip()
                if not line: # to skip any empty lines
                    continue
                    
                split_line = line.split("\t")
                
                if idx == 0:
                    col_names = split_line
                else:
                    data_lines.append(split_line)
                idx += 1
```

```python
df = pd.DataFrame(data_lines, columns=col_names) #create a pandas DataFrame from the collected data
df
```

```python
# This code was generated using AI. Prompt used: "Write a python script that convert specific DataFrame columns to numeric types in pandas, coercing any non-numeric values to NaN"
numeric_columns = [
    "CellViability.Mut.Rep1", "CellViability.Mut.Rep2", "CellViability.Mut.Rep3",
    "CellViability.WT.Rep1", "CellViability.WT.Rep2", "CellViability.WT.Rep3",
    "mRNA.Expression.Mut.Rep1", "mRNA.Expression.Mut.Rep2", "mRNA.Expression.Mut.Rep3",
    "mRNA.Expression.WT.Rep1", "mRNA.Expression.WT.Rep2", "mRNA.Expression.WT.Rep3",
    "Protein.Expression.Mut.Rep1", "Protein.Expression.Mut.Rep2", "Protein.Expression.Mut.Rep3",
    "Protein.Expression.WT.Rep1", "Protein.Expression.WT.Rep2", "Protein.Expression.WT.Rep3"
]

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Calculate mean values for each group
df["Viability.Mut.Mean"] = df[[
    "CellViability.Mut.Rep1", "CellViability.Mut.Rep2", "CellViability.Mut.Rep3"
]].mean(axis=1)

df["Viability.WT.Mean"] = df[[
    "CellViability.WT.Rep1", "CellViability.WT.Rep2", "CellViability.WT.Rep3"
]].mean(axis=1)

df["mRNA.Mut.Mean"] = df[[
    "mRNA.Expression.Mut.Rep1", "mRNA.Expression.Mut.Rep2", "mRNA.Expression.Mut.Rep3"
]].mean(axis=1)

df["mRNA.WT.Mean"] = df[[
    "mRNA.Expression.WT.Rep1", "mRNA.Expression.WT.Rep2", "mRNA.Expression.WT.Rep3"
]].mean(axis=1)

df["Protein.Mut.Mean"] = df[[
    "Protein.Expression.Mut.Rep1", "Protein.Expression.Mut.Rep2", "Protein.Expression.Mut.Rep3"
]].mean(axis=1)

df["Protein.WT.Mean"] = df[[
    "Protein.Expression.WT.Rep1", "Protein.Expression.WT.Rep2", "Protein.Expression.WT.Rep3"
]].mean(axis=1)
```

```python
# Check if mean was calculated correctly
df[["Viability.Mut.Mean", "mRNA.Mut.Mean", "Protein.Mut.Mean"]].head()
```

```python
# This code was generated using AI. Prompt used: "Write a python script that calculates fold changes"
# Avoid division by zero by replacing 0 with NaN

df["Viability_FoldChange"] = df["Viability.Mut.Mean"] / df["Viability.WT.Mean"].replace(0, np.nan)
df["mRNA_FoldChange"] = df["mRNA.Mut.Mean"] / df["mRNA.WT.Mean"].replace(0, np.nan)
df["Protein_FoldChange"] = df["Protein.Mut.Mean"] / df["Protein.WT.Mean"].replace(0, np.nan)
```

```python
# Log2 transform (symmetric for visualization). 
# This code was generated using AI. Prompt used: "Write a Python script to calculate fold changes between mutant and wild-type means for viability, mRNA, and protein data, replacing zeros in the denominator with NaN to avoid division by zero"
df["Viability_Log2FC"] = np.log2(df["Viability_FoldChange"])

df[[
    "Viability.Mut.Mean", "Viability.WT.Mean", "Viability_FoldChange", "Viability_Log2FC"
]].head()
```

```python
# Classify mutation type (substitution, insertion, deletion)

def classify_mutation_type_by_length(wt_seq, mut_seq):
    if len(mut_seq) > len(wt_seq):
        return "ins"
    elif len(mut_seq) < len(wt_seq):
        return "del"
    elif len(mut_seq) == len(wt_seq):
        return "sub"
    else:
        return "unknown"
```

```python
df["Mutation.Class"] = df.apply(
    lambda row: classify_mutation_type_by_length(row["WildType.Sequence"], row["Mutant.Sequence"]),
    axis=1
)
```

```python
# Check how many mutations are in each category
df["Mutation.Class"].value_counts()
```

```python
# This code was generated using AI. Prompt used: "Could you help me create a code which uses slicing and indexing to identify whether a mutation occurs in the promoter or coding region of WT and mutant sequences? If the promoter region is the first 1000 nucleotides"
# Identify whether the mutation is in the promoter region (first 1000 bases) or the coding region.

def mutation_region (wild_seq, mutant_seq) :
    if wild_seq [: 1000] != mutant_seq[: 1000] :
        return "Promoter"
    elif wild_seq != mutant_seq:
        return "Coding"
    else:
        return "No mutation"
        
df ["Mutation_Score"] = df.apply(lambda row: mutation_region (row["WildType.Sequence"], row ["Mutant.Sequence"]),
axis=1)

print(df [ ["Gene", "Mutation.Class", "Mutation_Score"]])
```

```python
df["mRNA_Log2FC"] = np.log2(df["mRNA_FoldChange"])
df["mRNA_Log2FC"] = df["mRNA_Log2FC"].replace([np.inf, -np.inf], np.nan)
```

```python
# This code was generated using AI. Prmpt used: "Generate Python code using matplotlib and seaborn to visualize mutation data. Use a subplot mosaic layout with a boxplot, scatter plot, and KDE plot. Assign custom colors for mutation classes ('sub', 'ins', 'del'), apply log scales to appropriate axes, and save the figure as a PNG. Add descriptive titles and a legend."

import matplotlib.pyplot as plt
import seaborn as sns

# Define custom colors for mutation types
palette = {"sub": "skyblue", "ins": "lightgreen", "del": "salmon"}

# Create mosaic layout: top row = box & scatter, bottom = full-width KDE
fig, ax = plt.subplot_mosaic(
    [["box", "scatter"],
     ["kde", "kde"]],
    figsize=(10, 10),
    layout="constrained"
)

# Boxplot
sns.boxplot(data=df, x="Mutation.Class", y="mRNA.WT.Mean", palette=palette, ax=ax["box"])
ax["box"].set_yscale("log")
ax["box"].set_title("mRNA WT Mean by Mutation")

# Scatter plot
sns.scatterplot(data=df, x="mRNA.Mut.Mean", y="Protein.Mut.Mean", hue="Mutation.Class", palette=palette, ax=ax["scatter"])
ax["scatter"].set_xscale("log")
ax["scatter"].set_yscale("log")
ax["scatter"].set_title("Protein vs mRNA (Mutant)")

# KDE plot
for mut, color in palette.items():
    sns.kdeplot(data=df[df["Mutation.Class"] == mut], x="Viability.Mut.Mean", ax=ax["kde"], label=mut, color=color)

ax["kde"].set_title("Viability Distribution (Mutant)")
ax["kde"].legend(title="Mutation Type")

plt.savefig("mutation_analysis_summary.png", dpi=300)
plt.show()
```

```python
# This code was generated using AI. Prompt used: "Write Python code using pandas and numpy to calculate mutation scores. Take log2 of protein fold change (handling zeros), drop rows with missing data, compute a cumulative mutation score from viability, mRNA, and protein log2FC values, then sort and display the top 5 scoring mutants with selected columns."

# Calculate Protein_Log2FC
df["Protein_Log2FC"] = np.log2(df["Protein_FoldChange"].replace(0, np.nan))

# Step 1: Drop any rows with missing expression or viability data
df_filtered = df.dropna(subset=[
    "Viability_Log2FC", "mRNA_Log2FC", "Protein_Log2FC"
]).copy()  # <- this ensures it's a true DataFrame copy


# Step 2: Create a scoring system (sum of all 3 log2FC values)
df_filtered["Mutation.Score"] = (
    df_filtered["Viability_Log2FC"] +
    df_filtered["mRNA_Log2FC"] +
    df_filtered["Protein_Log2FC"]
)



# Step 3: Sort by score (higher = better) and show top 5
top_5_mutants = df_filtered.sort_values(by="Mutation.Score", ascending=False).head(5)

# Step 4: Display only selected columns
top_5_mutants[[
    "Gene", "Mutation.Score", "Mutation.Class",
    "Viability_Log2FC", "mRNA_Log2FC", "Protein_Log2FC"
]]
```
